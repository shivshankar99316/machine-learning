```python
import numpy
print('Numpy ::', numpy.__version__)
import pandas
print('Pandas::',pandas.__version__)
import numpy as np
```

```python
a = np.arange(4).reshape((2,2))
a
```

```python
np.amin(a)
```

```python
# Find minimum along columnwise
np.amin(a,axis=0)
```

```python
# find minimum along row wise
np.amin(a,axis=1)
```

```python
a = np.array([[3,7,5],[8,4,3],[2,4,9]])
print ( 'Our array is:')
print(a)
print('\n')
```

```python
print(a)
```

```python
print ("\n")
```

```python
# minimum function
print ('Applying amin() function:rowwise')
print (np.amin(a,1) )
print ('\n')
```

```python
# we can find the minimum value within each column by specifying axis=0:
print ('Applying amin() function again : columnwise')
print (np.amin(a,0))
print ('\n')
```

```python
# Similarly we can find the  maximum value with in each row:
# maximum function
print ('Applying amax() function:')
print ( np.amax(a) )
print ('\n')
```

```python
print ('Applying amax() function again:')
print (np.amax(a, axis= 0))
```

```python
# The numpy.ptp () function return return the range (maximum-minimum) o values along on axis.
a = np.array([[3,7,5],[8,4,3],[2,4,9]])

print ('Our array is:')
print (a)
print('\n')
```

```python
print ('Applying ptp() function:')
print (np.ptp(a))
print('\n')
```

```python
print ('Applying ptp() function along axis 1: rowwise')
print (np.ptp(a, axis = 1))
print ('\n')
```

```python
print ('Applying ptp() function along axis 0:columnwise')
print (np.ptp(a, axis = 0))
```
